```python
import pandas as pd
import json
```

```python
MESSAGES_CSV = "messages.csv"
MEMBERS_CSV = "members.csv"
```

```python
messages_df = pd.read_csv("messages.csv")
members_df = pd.read_csv("members.csv")
```

```python
messages_df.head()
```

```python
# drop the Unnamed Column
members_df = members_df[['group_member_index', 'names', 'nicknames']]

# set index to member_index
members_df = members_df.set_index('group_member_index')
```

```python
print(members_df.shape)
members_df.head()
```

```python
# filter for user messages only
user_messages_df = messages_df[messages_df['member_index'] != "system"]

# drop the Unnamed Column
user_messages_df = user_messages_df[['member_index', 'group_message_index', 'text']]

# set index to member_index
user_messages_df = user_messages_df.set_index('member_index')
```

```python
print(user_messages_df.shape)
user_messages_df.head()
```

```python
user_messages_df.join(members_df)
```

```python
members_df.join(user_messages_df)
```

# https://stackoverflow.com/questions/22219004/grouping-rows-in-list-in-pandas-groupby

```python
user_messages_df.groupby('member_index').apply(lambda x: list(x['text'])).reset_index('member_index')
# members_df.join(user_messages_df, on=['group_member_index', 'member_index'], how="inner")
```

```python
aggregated_user_messages_df = user_messages_df.groupby('member_index').agg(list)
aggregated_user_messages_df.head()
```

```python
user_messages_df.groupby('member_index').agg(set)
```

```python
final_df = pd.concat([members_df, aggregated_user_messages_df])
```
